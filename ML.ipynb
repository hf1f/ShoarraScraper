```python
import pandas as pd

# Load datasets
poems_df = pd.read_json("all_poems_database_cleaned.json")
poets_df = pd.read_json("poets_database.json")

# Display shapes
print(f"Poems Dataset Shape: {poems_df.shape}")
print(f"Poets Dataset Shape: {poets_df.shape}")
```

```python
# Show structural info for poems
poems_df.info()

# Check missing values directly
poems_df.isnull().sum()
```

```python
# Preview first 5 rows with specific columns
poems_df[['poem_id', 'poet_name', 'title', 'era']].head()
```

```python
# Quick distribution of Eras
poems_df['era'].value_counts()
```

```python
# Number of poems per poet (Showing top 20)
poems_per_poet = poems_df['poet_name'].value_counts()

print("--- Top 20 Poets by Poem Count ---")
poems_per_poet.head(20)
```

```python
# Sub-themes distribution (e.g., أحزان, أفراح)
sub_theme_counts = poems_df['poem_sub_theme'].value_counts()

print("--- Poem Sub-Themes Distribution ---")
sub_theme_counts
```

```python
# Main themes distribution
theme_counts = poems_df['theme'].value_counts()

print("--- Main Themes Distribution ---")
theme_counts
```

```python
# Cross-tabulation between Era and Sub-Theme
pd.crosstab(poems_df['era'], poems_df['poem_sub_theme'])
```

```python
# Number of unique poets per era
poets_per_era = poems_df.groupby('era')['poet_name'].nunique().sort_values(ascending=False)

print("--- Number of Unique Poets per Era ---")
poets_per_era
```

```python
# Compare number of poems vs number of unique poets per era
era_stats = poems_df.groupby('era').agg(
    total_poems=('poem_id', 'count'),
    unique_poets=('poet_name', 'nunique')
).sort_values(by='total_poems', ascending=False)

era_stats
```
